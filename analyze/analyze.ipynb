```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
human = pd.read_csv('humanv2.csv') # 16
rl_1 = pd.read_csv('1rl.csv') # 14
rl_2 = pd.read_csv('2rl.csv') # 15
rl_3 = pd.read_csv('3rl.csv') # 16
```

```python
for i, frame in enumerate([human, rl_1, rl_2, rl_3]):
    grouped = frame.groupby(by=['time']).sum().reset_index()
    grouped['fuel'] = grouped.fuel.cumsum()
    plt.plot(grouped['time'], grouped['fuel'], label=f"{i} RL")
plt.legend(loc='best')
plt.show()
```

```python
for i, frame in enumerate([human, rl_1, rl_2, rl_3]):
    plt.bar(f"{i} RL", frame['speed'].mean())
    print(frame['speed'].mean())
plt.show()
```

```python
rls = rl_3[rl_3['id'].str.contains("rl")]
humans = rl_3[rl_3['id'].str.contains("human")]
```

```python
for i, frame in enumerate([humans, rls]):
    grouped = frame.groupby(by=['time']).mean().reset_index()
    grouped['fuel'] = grouped.fuel.cumsum()
    plt.plot(grouped['time'], grouped['fuel'], label=f"{i} RL")
plt.legend(loc='best')
plt.show()
```

```python
for i, frame in enumerate([humans, rls]):
    plt.bar(f"{bool(i)} RL", frame['speed'].mean())
    print(frame['speed'].mean())
plt.show()
```
